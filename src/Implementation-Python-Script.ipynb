```python
import networkx as nx

# Create graph
G = nx.DiGraph()

# Add nodes
G.add_node('Source')
G.add_node('Sink')
G.add_node('FH205_AP')
G.add_node('FH207_AP')

# Add edges with capacities
G.add_edge('Source', 'FH205_AP', capacity=1000)
G.add_edge('Source', 'FH207_AP', capacity=1000)
G.add_edge('FH205_AP', 'Sink', capacity=500)
G.add_edge('FH207_AP', 'Sink', capacity=500)

# Compute max flow
flow_value, flow_dict = nx.maximum_flow(G, 'Source', 'Sink')

# Print results
print("=" * 50)
print("NETWORK FLOW ANALYSIS - FEDERICO HALL")
print("=" * 50)
print(f"Max flow: {flow_value} Mbps")
print(f"Required: 3350 Mbps")
print(f"Students served: {flow_value / 5} out of 670")
print()

if flow_value < 3350:
    print("RESULT: NOT ENOUGH")
    print(f"Short by: {3350 - flow_value} Mbps")
else:
    print("RESULT: ENOUGH")
```
